## ëª¨ë“ˆ import

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

## ë°ì´í„° ì¤€ë¹„

```python
from sklearn.datasets import load_digits

# ë°ì´í„° ë¡œë“œ
digits = load_digits()
print(dir(digits))
```

```python
# ë°ì´í„° í™•ì¸
digits_data = digits.data
print(digits_data.shape) # ë°ì´í„° í¬ê¸°
print(digits_data[0])

# print(digits.feature_names) # featuresëŠ” í”½ì…€ê°’

digits_label = digits.target
print(digits.target_names)
```

```python
from collections import Counter
print(Counter(digits_label))
```

```python
print(digits.DESCR)
```

## train, test ë°ì´í„° ë¶„ë¦¬

```python
X_train, X_test, y_train, y_test = train_test_split(digits_data, 
                                                    digits_label,
                                                   test_size=0.2,
                                                   random_state=7)
print('X_train ê°œìˆ˜: ', len(X_train), ', X_test ê°œìˆ˜: ', len(X_test))
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

## ëª¨ë¸ í•™ìŠµ ë° í‰ê°€

### Decision Tree



```python
# ëª¨ë¸ í•™ìŠµ
decision_tree = DecisionTreeClassifier(random_state=32)
decision_tree.fit(X_train, y_train)
y_pred = decision_tree.predict(X_test)

# ëª¨ë¸ í‰ê°€
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

### Random Forest 

```python
# ëª¨ë¸ í•™ìŠµ
random_forest = RandomForestClassifier(random_state=32)
random_forest.fit(X_train, y_train)
y_pred = random_forest.predict(X_test)

# ëª¨ë¸ í‰ê°€
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

### SVM

```python
# ëª¨ë¸ í•™ìŠµ
svm_model = svm.SVC(random_state=32)
svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)

# ëª¨ë¸ í‰ê°€
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

### SGD Classifier

```python
# ëª¨ë¸ í•™ìŠµ
sgd_model = SGDClassifier(random_state=32)
sgd_model.fit(X_train, y_train)
y_pred = sgd_model.predict(X_test)

# ëª¨ë¸ í‰ê°€
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

### Logistic Regression

```python
# ëª¨ë¸ í•™ìŠµ
logistic_model = LogisticRegression(random_state=32, max_iter=2700)
logistic_model.fit(X_train, y_train)
y_pred = logistic_model.predict(X_test)

# ëª¨ë¸ í‰ê°€
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
